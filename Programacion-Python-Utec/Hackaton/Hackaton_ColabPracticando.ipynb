```python
df2 = df[["id","host_id", "listing_url","host_name", "host_is_superhost", "name", "neighbourhood", "neighbourhood_cleansed", "host_response_rate", "latitude", "longitude", "room_type", "price", "number_of_reviews", "review_scores_rating" ,"accommodates","bathrooms" ]].copy()
```

```python
df2.head()
```

```python
df2.dropna(inplace=True)
df2
```

```python
a='1234$'
print(a.replace("$",""))
```

```python
#Limpiamos algunas variables como price y host_response_rate
def limpiar(a):
  return str(a).replace("$","").replace(",","")

df2["price"] = df2["price"].apply(limpiar)#replace("[$,]", "")
```

```python
# df2["host_response_rate"] = df2["host_response_rate"].str.replace("%","")
```

```python
#Convertimos el precio y host_response_rate a float
df2['price'] = df2['price'].astype(float)

df2["host_response_rate"] = df2['host_response_rate'].astype(float)
```
