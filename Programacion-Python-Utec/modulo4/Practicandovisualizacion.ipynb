```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,3,5,7,11]

plt.plot(x,y, marker='o', color='#b2df8a', linestyle='--', label='linea') #color: se extrajo codigo de paleta de colores colorbrewer2.org , sino se coloca: blue, red, green.
plt.plot(y,x, marker='x', color='#1f78b4', linestyle='--', label='linea2')

plt.title('Mi primer grafico')
plt.xlabel('Edad')
plt.ylabel('Nota')

plt.legend() #Este reconoce el "label"= linea
plt.grid(True) #crea una grilla

plt.show() #es como el print
```
